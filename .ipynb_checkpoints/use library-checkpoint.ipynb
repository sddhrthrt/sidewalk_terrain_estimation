```python
gpxfiles = !ls files/gpx*
accfiles = !ls files/acc*
```

```python
import re
import shutil
```

```python
for file in gpxfiles:
  gts = int(re.findall('\d+', file)[0])
  count = 0
  tdiff = 20
  match = ""
  for afile in accfiles:
    ats = int(re.findall('\d+', afile)[0])
    if abs(gts-ats)<10:
      match = afile
      count+=1
      if abs(gts-ats) < tdiff:
        match = afile
  if match:
      ts = re.findall('\d+', match)[0]
      newts = re.findall('\d+', file)[0]
      if not ts == newts:
        newname = match.replace(ts, newts)
        shutil.copy(match, newname)
```
