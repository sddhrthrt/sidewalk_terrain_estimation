```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.signal import butter, lfilter, freqz
import scipy
```

```python
def drawAccelerometerPlots(df, tags, cols):
    n = len(cols)
    print(n)
    fig, axes = plt.subplots(nrows=1, ncols=n)
    if n==1: axes=[axes,]
    for i in range(n):
      df.reset_index().plot(x='tsr', y=cols[i], ax=axes[i], figsize=(18, 21/n))
      ymin, ymax = axes[i].get_ylim()
      axes[i].vlines(x=tags['tsr']*1000, ymin=ymin, ymax=ymax)
    plt.show()

def readFilesAndPlotAccelerometer(filelist):
    dfs = {}
    
    for fn in filelist:
        dfi = pd.read_csv(fn,
                     header=None,
                     names=['ts', 'a_x', 'a_y', 'a_z']
                    )
        tags_df = pd.read_csv(fn.replace("accel_raw", "tags"),
                              header=None,
                              names=['ts', 'tag']
                             )
        dfi['tag'] = ''
        dfi['tag_m'] = 0
        tags = ['TAG_1', 'TAG_2', 'TAG_3', 'TAG_4']
        for i, r in tags_df.iterrows():
          range_i = ((dfi['ts']/1000).astype('int')==(r['ts']))
          dfi['tag'][range_i] = r['tag']
          dfi['tag_m'][range_i] = tags.index(r['tag'])+1 
        dfi['tsr'] = dfi['ts']-dfi['ts'][0]
        tags_df['tsr'] = tags_df['ts']-(dfi['ts']/1000).astype('int')[0]
        dfs[fn] = dfi
        dfs[fn+'_t'] = tags_df
        drawAccelerometerPlots(dfi, tags_df, ['a_x', 'a_y', 'a_z'])
    return dfs

def drawGyroPlots(df):
    fig, axes = plt.subplots(nrows=1, ncols=3)
    df.reset_index().plot(x='tsr', y='gx', ax=axes[0], figsize=(18, 7))
    df.reset_index().plot(x='tsr', y='gy', ax=axes[1], figsize=(18, 7))
    df.reset_index().plot(x='tsr', y='gz', ax=axes[2], figsize=(18, 7))

def readFilesAndPlotGyro(filelist):
    dfs = {}
    for fn in filelist:
        dfi = pd.read_csv(fn,
                     header=None,
                     names=['ts'] + ['gx', 'gy', 'gz'],
                    )
        dfi['tsr'] = dfi['ts']-dfi['ts'][0]
        dfs[fn] = dfi
        drawGyroPlots(dfi)
    plt.show()
    return dfs
```

### integration

```python
def acumtrapzd(df, col, xcol):
    a = df[col]
    t = df[xcol]
    v = scipy.integrate.cumtrapz(a, t, initial=0)
    d = scipy.integrate.cumtrapz(v, t, initial=0)
    return v, d
```

## Shortest paths

###  maybe limit the number of roads in this 

```python
lt, rb = [max(dots[:,0]), min(dots[:,1])], [min(dots[:,0]), max(dots[:,1])]
# widen
latheight = rb[0]-lt[0]
lngwidth = rb[1]-lt[1]
latinc = latheight*.3
lnginc = lngwidth*.3
lt[0] -= latinc
rb[0] += latinc
lt[1] -= lnginc
rb[1] += lnginc
```

```python
lat_bounds_1 = np.logical_and((P1[:,0]<lt[0]), (P1[:,0]>rb[0]))
lng_bounds_1 = np.logical_and((P1[:,1]>lt[1]), (P1[:,1]<rb[1]))
lat_bounds_2 = np.logical_and((P2[:,0]<lt[0]), (P2[:,0]>rb[0]))
lng_bounds_2 = np.logical_and((P2[:,1]>lt[1]), (P2[:,1]<rb[1]))

p1_bounds = np.logical_and(lat_bounds_1, lng_bounds_1) 
p2_bounds = np.logical_and(lat_bounds_2, lng_bounds_2)
filter_area = p1_bounds + p2_bounds 
```

```python
P1i = P1[filter_area]
P2i = P2[filter_area]
```

```python
def hashize(t):
    return (t*1e7).astype(int)
```

```python
ip1, ip2 = hashize(P1i), hashize(P2i)
```

```python
ipl1 = list(map(tuple, ip1))
ipl2 = list(map(tuple, ip2))
```

```python
ipl = ipl1 + ipl2
```

```python
distdict = {t: {} for t in set(ipl)}
# distdict[point1][point2] = distance from point1 to point2 on sidewalk.
```

```python
distances = great_circle_dist(P1i, P2i)
```

```python
# direct sidewalk distances.
for i in range(len(P1i)):
    p1 = tuple(hashize(P1i[i]))
    p2 = tuple(hashize(P2i[i]))
    distance = distances[i]
    distdict[p1][p2] = distance
```

```python
# reverse sidewalk distances.
for k, v in distdict.items():
    for k2, v2 in v.items():
        if not k in distdict[k2]:
            distdict[k2][k] = v2
```

```python
rhs = np.vstack((P1i, P2i))
alt_distances = great_circle_dist_from_dots(*list(map(np.radians, [P1i, P2i, rhs])))
# ON HOLD
# for every point, we'll add <10m roads as-is
# for 10 - 30, we'll double the distance because it's not exactly sidewalk
```

```python
for i in range(len(P1i)):
    res = great_circle_dist(P1i[[i]], P2i[i+1:])
    k1 = tuple(hashize(P1i[i]))
    for j in range(i+1, len(P1i)):
        k2 = tuple(hashize(P2i[j]))
        if res[j-i-1] < 0.01:
            save_as = res[j-i-1]
            distdict[k1][k2] = save_as
            distdict[k2][k1] = save_as
        elif res[j-i-1] < 0.03:
            save_as = res[j-i-1]*3
            distdict[k1][k2] = save_as
            distdict[k2][k1] = save_as
```
